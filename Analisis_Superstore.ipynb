```python
import pandas as pd

# Load the dataset into a Pandas DataFrame
df = pd.read_csv('Superstore.csv', encoding='latin-1')

print(df.shape)

# Display information about the DataFrame (column types, non-null values, etc.)
print(df.info())

# Display descriptive statistics of the numerical columns
print(df.describe())
```

```python
df
```

# Data Cleansing

```python
# Hapus baris duplikat
df.drop_duplicates(inplace=True)

# Konversi tanggal
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Cek missing values
print(df.isnull().sum())
```

```python
df.shape
```

```python
# Bulan & Tahun
df['Order Month'] = df['Order Date'].dt.to_period('M')

# Margin Profit
df['Profit Margin (%)'] = (df['Profit'] / df['Sales']) * 100
```

```python
df
```

# Kategori produk vs profit di setiap region

```python
pivot1 = df.groupby(['Region', 'Category'])['Profit'].sum().reset_index()
pivot1 = pivot1.sort_values(['Region', 'Profit'], ascending=[True, False])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(data=pivot1, x='Category', y='Profit', hue='Region')
plt.title('Profit per Category per Region')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Apakah diskon meningkatkan penjualan atau menurunkan profit?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot
sns.scatterplot(data=df, x='Discount', y='Profit')
plt.title("Hubungan Diskon vs Profit")
plt.show()

# Korelasi
correlation = df[['Discount', 'Profit']].corr()
print(correlation)
```

```python
# prompt: korelasi matrix discount dan profit

import matplotlib.pyplot as plt
# Visualisasi Korelasi
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Korelasi antara Diskon dan Profit')
plt.show()
```

# Tren penjualan dan profit bulanan



```python
monthly_trend = df.groupby('Order Month')[['Sales', 'Profit']].sum().reset_index()

# Plot tren
monthly_trend['Order Month'] = monthly_trend['Order Month'].astype(str)
monthly_trend.plot(x='Order Month', y=['Sales', 'Profit'], figsize=(12,6), marker='o')
```

# Wilayah dengan kerugian terbesar


```python
# Total profit per state
loss_region = df.groupby(['State'])['Profit'].sum().reset_index()
loss_region = loss_region.sort_values('Profit')  # dari yang paling rugi

print(loss_region.head(5))  # Top 5 negara bagian paling merugi
```

# Segmen pelanggan paling menguntungkan dan produk favoritnya


```python
# Profit per segment
segment_profit = df.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
print(segment_profit)

print('*****************************')

# Produk terlaris per segmen
top_products = df.groupby(['Segment', 'Product Name'])['Sales'].sum().reset_index()
top_products = top_products.sort_values(['Segment', 'Sales'], ascending=[True, False])
print(top_products.head(10))
```
